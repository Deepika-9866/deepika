```python
#strings-collection of characters
''' 
string can be defined in single,doubleand triple quotes(apostrophe)
lower,upper,endswith,startswith,replace,find,index,count,remove suffix or prefix,split,rstrip,strip,lstrip
'''
a = "good morning"
print(a.upper())
print(a.lower())
print(a.endswith("morning"))
print(a.startswith("g"))
```

```python
b = "hello everyone"
print(b.replace("everyone","all"))
print(b.index("hello"))
print(b.find("hii"))
#in find if you ask any string which is not there the output is -1
print(b.count('e'))
print(b.removeprefix('hello'))
print(b.removesuffix('everyone'))
```
